```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome('/usr/bin/chromedriver')

url = "https://kumparan.com/trending"

driver.get(url)
```

```python
import pandas as pd

df = pd.DataFrame(columns=['Judul', 'Penulis', 'JumlahSuka', 'JumlahKomentar']) #Main Data Frame

posts = driver.find_elements_by_css_selector('[data-qa-id="news-item"]')

for post in posts:
    title   = post.find_element_by_css_selector('[data-qa-id="title"]').text
    author  = post.find_element_by_css_selector('[data-qa-id="author-name"]').text
    like    = post.find_element_by_css_selector('[data-qa-id="like-button-wrapper"]').text
    like    = like if like else 0 #Fill like with 0 if empty
    comment = post.find_element_by_css_selector('[data-qa-id="comment-button-wrapper"]').text
    comment = comment if comment else 0 #Fill comment with 0 if empty
    newDf   = {
        "Judul" : title,
        "Penulis" : author,
        "JumlahSuka" : like,
        "JumlahKomentar" : comment
    }
    df = df.append(newDf, ignore_index=True)
    
print(df)
```
